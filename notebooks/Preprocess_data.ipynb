```python
import pandas as pd
import pathlib
import numpy as np
```

```python
BASE_DIR = pathlib.Path().resolve().parent
DATA_DIR = BASE_DIR/"data"
CSV_PATH = DATA_DIR/"raw_csv.csv"
PROCESSED_CSV_PATH = DATA_DIR/"processed_csv.csv"
METADATA_PATH = DATA_DIR/"metadata.json"
```

```python
df = pd.read_csv(CSV_PATH)
df
```

```python
one_hot_df = pd.get_dummies(df, columns=['protocol_type', 'service', 'flag', 'labels'], dtype=float)
column_names = one_hot_df.columns
one_hot_df
```

```python
one_hot_df.columns = np.arange(len(column_names))
```

```python
def remove_const_features(df):
    for i in range(len(df.columns)):
        if df[i].nunique() == 1:
            df.drop([i], axis=1, inplace=True)
```

```python
remove_const_features(one_hot_df)
```

```python
one_hot_df.columns = np.arange(len(one_hot_df.columns))
```

```python
def normalise_data(df):
  for i in range(len(df.columns)):
    if df[i].nunique() > 2:
       min = df[i].min()
       max = df[i].max()
       df[i] = df[i].apply(lambda x: (x-min)/(max-min))
```

```python
normalise_data(one_hot_df)
```

```python
one_hot_df.to_csv(PROCESSED_CSV_PATH, index=False)
```
